```python
import sys
import os
sys.path.append(os.path.abspath('../src'))

import pandas as pd
import numpy as np
from lookalike_model import LookalikeModel
```

```python
customers_df = pd.read_csv('../data/raw/Customers.csv')
products_df = pd.read_csv('../data/raw/Products.csv')
transactions_df = pd.read_csv('../data/raw/Transactions.csv')

print("Data Loading Summary:")
print(f"Customers: {len(customers_df)} records")
print(f"Products: {len(products_df)} records")
print(f"Transactions: {len(transactions_df)} records")
```

```python
model = LookalikeModel(customers_df, transactions_df, products_df)

# Generate recommendations for first 20 customers
results = []
for i in range(1, 21):
    customer_id = f'C{i:04d}'
    similar_customers, scores = model.find_similar_customers(customer_id)
    results.append({
        'source_customer_id': customer_id,
        'lookalike1_id': similar_customers[0],
        'score1': scores[0],
        'lookalike2_id': similar_customers[1],
        'score2': scores[1],
        'lookalike3_id': similar_customers[2],
        'score3': scores[2]
    })

# Create DataFrame and save results
results_df = pd.DataFrame(results)
results_df.to_csv('../output/FirstName_LastName_Lookalike.csv', index=False)
```

```python
# notebooks/FirstName_LastName_Lookalike.ipynb

# Add this code after model implementation:
def generate_lookalike_results():
    results = []
    for i in range(1, 21):
        customer_id = f'C{i:04d}'
        similar_customers, scores = model.find_similar_customers(customer_id)
        
        # Format for CSV
        result = {
            'source_customer': customer_id,
            'lookalike_customers': [
                {'customer_id': cust, 'similarity_score': score}
                for cust, score in zip(similar_customers, scores)
            ]
        }
        results.append(result)
    
    return pd.DataFrame(results)

# Generate and save results
lookalike_results = generate_lookalike_results()
lookalike_results.to_csv('../output/FirstName_LastName_Lookalike.csv', index=False)
```

```python
# notebooks/FirstName_LastName_Lookalike.ipynb

def generate_top20_recommendations():
    results = []
    for i in range(1, 21):
        customer_id = f'C{i:04d}'
        similar_customers, scores = model.find_similar_customers(customer_id, n=3)
        
        results.append({
            'source_customer': customer_id,
            'recommendations': [
                {'customer_id': cust, 'similarity_score': float(score)}
                for cust, score in zip(similar_customers, scores)
            ]
        })
    
    return pd.DataFrame(results)

# Generate and save results
top20_results = generate_top20_recommendations()
top20_results.to_csv('../output/FirstName_LastName_Lookalike.csv', index=False)
```
